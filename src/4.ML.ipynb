__MODELO RANDOM FOREST__

Entrenamos nuestro modelo con un modelo de casificaicon Random Forest

```python
import pandas as pd

train_data = pd.read_csv("../data/processed/clean_train.csv")
test_data = pd.read_csv("../data/processed/clean_test.csv")

train_data.head()
```

```python
X_train = train_data.drop(["Grave"], axis = 1)
y_train = train_data["Grave"]
X_test = test_data.drop(["Grave"], axis = 1)
y_test = test_data["Grave"]
X_test.head(5)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 40, random_state = 5)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

Hemos obtenido un buen accuracy , incluso mejor que entrenando el modeo con un modelo de Regresion Logistica que podemos ver el el archico 4.1 M.L.ipynb
Procedemos a guardar el modelo

```python

import pickle

ruta_guardado = '../models/randomforestmodel.pkl'
with open(ruta_guardado, 'wb') as archivo:
    pickle.dump(model, archivo)
```
